```python
# استدعاء المكتبات 
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
# سحب البيانات واستعراض جزء منها
data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/dataset.csv")
data.head(50)
```

```python
# التحقق من مجموع القيم غير الموجودة في مجموعة البيانات
data.isnull().sum()
```

```python
# التحقق من عدد عينات النصوص لكل لغة
data["language"].value_counts()
```

```python
# تحضير البيانات 
x = np.array(data["Text"])     # ميزة التدريب
y = np.array(data["language"]) # الهدف

# يستخدم لتحويل النص إلى متجه بناء على عدد مرات ظهور الكلمات (التكرار) في النص
cv = CountVectorizer()
X = cv.fit_transform(x)

# تقسيم البيانات إلى بيانات تدريب و اختبار
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.33, 
                                                    random_state=42)
```

```python
# تجهيز النموذج وموائمته مع بيانات التدريب 
model = MultinomialNB()
model.fit(X_train,y_train)

# تحديد دقة النموذج من خلال اختباره على بيانات الاختبار
score = model.score(X_test,y_test)
print("Model's score: ", score)
```

```python
user = input("Enter a Text: ")
data = cv.transform([user]).toarray()
output = model.predict(data)
print(output)
```
