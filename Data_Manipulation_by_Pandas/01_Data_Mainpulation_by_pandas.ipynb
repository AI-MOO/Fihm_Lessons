## Section 01: Transforming DataFrames | تحويل أطر البيانات


### `1.1 Inspecting a DataFrame | فحص إطار البيانات`


```python
# pandas استدعاء مكتبة
import pandas as pd
from IPython.display import display


# سحب البيانات واستعراضها على هيئة إطار بيانات
diamonds = pd.read_csv('Datasets/diamonds.csv')

# عرض إطار البيانات
display(diamonds.head())

# الحصول على معلومات إطار البيانات
print(diamonds.shape)
print(diamonds.info())
```

### `1.2 statistical and descriptive explanation of data | الشرح الوصفي والاحصائي للبيانات`


- carat: وزن الماسة بالقيراط ( الغرام الواحد = 5 قراريط.)
- cut: جودة القطع (مقبولة، جيدة ، جيدة جدًا ، ممتازة ، مثالية) | (Fair, Good, Very Good, Premium, Ideal)
- color: لون الماس ، من J (الأسوأ) إلى D (الأفضل)
- clarity: الوضوح قياس مدى وضوح الماس [I1 (الأسوأ) ، SI2 ، SI1 ، VS2 ، VS1 ، VVS2 ، VVS1 ، IF (الأفضل)]
- depth: النسبة المئوية للعمق الكلي = z / mean(x, y) = 2 * z / (x + y)
- table: عرض المنضدة العليا للماسة بالنسبة إلى أوسع نقطة
- price: السعر بالدولار الأمريكي.
- x: الطول بالملليمتر
- y: العرض بالملليمتر
- z: العمق بالملليمتر

```python
# الشرح الاحصائي للبيانات الرقمية
display(diamonds.describe())
```

### `1.3 DataFrame Parts | أجزاء إطار البيانات`


```python
# Numpy استخراج القيم على هيئة مصفوفة 
print(diamonds.values)
print(15 * '---')

# استخراج صف واحد فقط 
print(diamonds.values[0])
print(15 * '---')

# استخراج قيمة من عمود
print(diamonds.values[0][1])
```

```python
# طباعة أسماء الأعمدة
print(diamonds.columns)
print(15 * "---")

# طباعة نطاق الفهرسة
print(diamonds.index)
```

### `1.4 Rows Sorting | ترتيب الصفوف`

```python
#  ترتيب الصفوف وفقًا لسمة السعر بشكل تصاعدي
diamonds_price = diamonds.sort_values("price")

# عرض أول 10 صفوف
display(diamonds_price.head(10))
```

```python
#  ترتيب الصفوف وفقًا لسمة السعر بشكل تنازلي
diamonds_price = diamonds.sort_values("price", ascending = False)

# عرض أول 10 صفوف
display(diamonds_price.head(10))
```

```python
#  ترتيب الصفوف وفقًا لسمتي السعر ونوع القطع  

diamonds_price_cut = diamonds.sort_values(["cut", "price"] , ascending = [True, False])
display(diamonds_price_cut)
```

### `1.5 Subsetting by columns | التجزئة من خلال الأعمدة`

```python
#  التجزئة من خلال اختار عامود واحد فقط (سمة)  
cut = diamonds["cut"]

# عرض الاجتزاء
display(cut.head(7))

# فاصل
print(20 * "---")

# التجزئة من خلال اختار أكثر من عمود (عدة سمات) 
multiple_attributes = diamonds[["cut", "clarity", "color"]]

# عرض الاجتزاء 
display(multiple_attributes.head(7))
```

### `1.6 Subsetting by rows | التجزئة من خلال الصفوف`

```python
# التجزئة من خلال ترشيح قيمة نصية محددة
G_diamonds = diamonds[diamonds["color"] == "D"]

# عرض النتيجة
display(G_diamonds)
```

```python
# التجزئة من خلال ترشيح الصفوف من خلال عدة قيم نصية 

best_cut_list = ["Premium", "Ideal"]
best_cut_diamonds = diamonds[diamonds["cut"].isin(best_cut_list)]

# عرض النتيجة
display(best_cut_diamonds)
```

```python
# ترشيح إطار البيانات ليتضمن الماسات ذات العمق الكلي الأكثر من 65
dia_depth_more_65 = diamonds[(diamonds["depth"] > 65)]

# عرض النتيجة
display(dia_depth_more_65)
```

### `1.7 Adding new columns to DataFrame | إضافة أعمدة (سمات) جديدة لإطار البيانات`

```python
# إضافة عمود جديد إلى إطار البيانات
diamonds["volume"] = diamonds["x"] * diamonds["y"] * diamonds["z"]

# عرض النتيجة
display(diamonds.head(10))
```
