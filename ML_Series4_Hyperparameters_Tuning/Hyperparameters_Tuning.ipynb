## `لماذا نحتاج ضبط عوامل التدريب؟`

`لفهم الهدف من ضبط عوامل التدريب علينا فهم أسباب الخطأ في نماذج تعلم الآلة أولا. حيث أن نماذج الآلة لا تصيب التوقع دائما، وعندما تخطئ يكون الخطأ إحدى الأسباب الثلاث التالية`

`
1 - Unavoidable randomness
`

`
هي الحقيقة أن العشوائية سوف تبقى دائما موجودة في البيانات ولكن تختلف من بيانات إلى أخرى. وهو خطأ لا يمكن تجنبه
`

`
2 - Bias is high 
`

`
أن يكون النموذج منحاز بشكل عالي وهو ما يحصل عند عدم تدريب النموذج لفترة كافية أو عند استخدام بيانات منحازة 
`

`
3 - Varaince is high 
`

`
أن يكون النموذج متابين بشكل عالي وهو ما يحصل عند تدريب النموذج بشكل زائد عن اللزوم.  
`

`
لتجنب تدريب النموذج زيادة عن اللزم أو عدم تدريبه بالشكل كافي. دعونا ننظر إلى مراحل تدريب النموذج كالتالي
`

### `First Stage: Underfit Model`




`
عند بداية تدريب النموذج يكون النموذج ذو انحياز عالي
`

`High bias`

`
لقلة كمية التدريب مثل التدرب على عشرة صفوف فقط، ولذلك تكون توقعات النموذج خاطئة ومتكررة 
`


### `Second Stage: Balanced Model`

`
في هذه المرحلة نقوم بتقليل أثر الانحياز العالي عن طريق الاستمرار في تدريب النموذج من أجل خفض الانحياز ورفع التباين مما يؤدي إلى رفعة دقة النموذج
`


### `Third Stage: Overfit Model`

`
 ولكن إذا استمرينا في التدريب بعد حد معين، يصبح النموذج معقد`

`complex`

`فيصبح ذو تباين عالي جدًا, مما يجعل النموذج يخطئ مرة أخرى. لذلك إن عملية الاستمرار في تدريب النموذج له أثر عكسي على الدقة، ومن خلال ضبط عوامل التدريب نستطيع تحديد متى يقف التدريب لتجنب جعل النموذج ذو تباين عالي
`


____
`الخلاصة: إن تدريب النموذج يقلل من الانحياز`

`Bias`

`مما يرفع من دقة النموذج ولكنه أيضا يرفع التباين`

`variance`

`رفع التباين مقبول إلى حد ما، ولكن إذا تجاوزنا هذا الحد تصبح قيمة التباين جدًا عالية مما يجعل النموذج يخطئ من جديد. هدفنا هو استخدام ضبط عوامل التدريب لوقف التدريب عند الحد المناسب الذي يضمن أقل انحياز ممكن مقابل رفع التباين بشكل معقول`

## `الانحياز ضد التباين - Bias vs Variance`

<img src="https://i.ibb.co/V9zn3Mk/Screen-Shot-2023-01-16-at-03-01-46.png" alt="Bias vs Variance" border="0" width = 350>

`حين النظر إلى الرسمة أعلاه نلاحظ أربعة نماذج يحاولون توقع مكان النقطة السوداء التي بالمنتصف`
______
`النموذج الأول هو نموذج تدرب على كمية صفوف قليلة جدًا أو على بيانات منحازة بالأصل ولذلك نجده يقوم بنفس التوقع الخاطئ مرارا وتكرارًا. يسمى هذا النموذج بالنموذج المنحاز `

`Biased model or underfit model`
________
` النموذج الثاني هو نموذج تم تدريبه على بيانات ذات جودة مناسبة وتم إيقاف تدريبه بالوقت المناسب. ولذلك فإن انحيازه وتبيانه قليلين. يسمي هذا النموذج بالنموذج المتزن`

`Balanced model`
_______
`النموذج الثالث هو نموذج تم تدريبه على بيانات ذات جودة عالية ولكن تم تدريبه زيادة عن اللزوم مما جعله نموذج مشبع وأصبحت أخطاؤه كثيره. ولذلك وعلى الرغم من أن هذا النموذج ذو انحياز قليل إلا أن تباينه عالي. يسمى هذا النموذج بالنموذج المعقد`

`Complex model or overfit model`
__________
`النموذج الرابعة هو نموذج تم تدريبه بشكل كثيف على بيانات ذات جودة منخفضة. مما جعل تباينه وانحيازه عاليين`
_______
`نلاحظ أن أفضل نموذج من ناحية الأداء هو النموذج الثاني حيث أننا استخدمنا بيانات ذو جودة مناسبة وقمنا بتدريب النموذج بشكل كافي، أي أن التدريب لم يكون لفترة قصيرة فيصبح النموذج منحاز ولم يكن لفترة طويلة فيصبح النموذج متباين`

`الخلاصة: النموذج المناسب هو النموذج الذي يوازن ما بين الانحياز والتباين`

`A balanced model is a model that balances between bias and variance`

## `Step One: Defining Auxiliaries functions - الخطوة الأولى: تعريف الوظائف المساعدة`

`سوف نقوم بتعريف وظيفة تقوم بتقييم أداء النموذج باستخدام المعايير التالية`

`Accuracy, Recall, Precision, F1-score`

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
def evaluate_metrics(yt, yp):
    results_pos = {}
    results_pos['accuracy'] = accuracy_score(yt, yp)
    precision, recall, f_beta, _ = precision_recall_fscore_support(yt, yp, average='binary')
    results_pos['recall'] = recall
    results_pos['precision'] = precision
    results_pos['f1score'] = f_beta
    return results_pos
```

 `سوف نقوم أيضا بتعريف وظيفة تقوم برسم شجرة القرار بعدما يتم تدريبها`

```python
from sklearn import tree
import matplotlib.pyplot as plt
def plot_decision_tree(model, feature_names):
    plt.subplots(figsize=(25, 20)) 
    tree.plot_tree(model, 
                       feature_names=feature_names,  
                       filled=True)
    plt.show()
```

`وسوف نقوم أيضا بتثبيت الحالة العشوائية في النماذج. والسبب بذلك إننا حينما نقوم باختيار صفوف الاختبار وصفوف التدريب فإننا نختار الصفوف بشكل عشوائي حسب قيمة الحالة العشوائية و التي تتغير في كل مرة. ولضمان إننا سوف نختبر النماذج المختلفة باستخدام نفس الصفوف نقوم بتوحيد الحالة العشوائية 
`

`
random_state
`

```python
rs = 9981 # Setting random state to 9981
```

`ملاحظة: اذا قمت باستخدام 9981 كرقم الحالة العشوائية فسوف تحصل على نفس النتائج التي لدينا. وتستطيع أيضا تجربة أرقام أخرى لرؤية أثرها`

## `Step Two: Built A Decision Tree Model - الخطوة الثانية: بناء نموذج شجرة القرار `


`إن أحد مزايا خوارزمية شجرة القرار هو أننا لا نحتاج إلى إعادة وزن الأعمدة`

`Scaling`

`ولذلك سوف نقوم بفصل البيانات مباشرة`

```python
X = df.drop('Class', axis = 1) # Training features
y = df.Class # Target features 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state = rs)
```

`الآن سوف نقوم ببناء نموذج شجرة قرار بدون ضبط عوامل التدريب والتحقق من أداء النموذج`

```python
from sklearn.tree import DecisionTreeClassifier
plain_model = DecisionTreeClassifier(random_state=rs)
plain_model.fit(X_train, y_train)
preds = plain_model.predict(X_test)
```

```python
evaluate_metrics(y_test, preds)
```

`نلاحظ التالي: لقد حصلنا على دقة تساوي 92.4% وعلى قيمة إف-ون تساوي 89.6% باستخدام نموذج شجرة القرار بدون ضبط عوامل التدريب`

____
` فلنقم برسم الشجرة لمعرفة كيفية شكلها (كيف تشكلت بناء على التدريب على البيانات المزودة)`

```python
feature_names = X.columns.values #Taking columns names 
plot_decision_tree(plain_model, feature_names)
```

`.هناك ثلاث ألون لإوراق الشجرة كالتالي`


`
الأوراق البرتقالية هي أوراق يكون معظم الصفوف فيها مصنفة على أنها 0 أي أنها أورام حميدة
`

`
الأوراق الزرقاء هي الأوراق التي يكون معظم الصفوف فيها مصنفة على أنها 1 أي أورام حميدة 
`


`
أخيرا الأوراق البيضاء هي الأوراق التي تساوت عدد عينات الأورام الحميدة والخبيثة فيها
`


`نلاحظ أن شجرة القرار التي نتجت لدينا هي شجرة معقدة نوعًا ما ولديها العديد من التفرعات. تذكر إنه كلما كبرت الشجرة كل ما زاد تعقيدها وأصبحت ذات تباين عالي جدًا`

`Longer decision trees tend to overfit and have high variance`

## `Step Three: Tune Decision Tree Hyperparameters - الخطوة الثالثة: ضبط عوامل تدريب شجرة القرار `

`يوجد العديد من عوامل التدريب لشجرة القرار. وللإطلاع عليها قم بتجربة الكود التالي`

```python
# DecisionTreeClassifier?
```

`في درس اليوم، سوف نقوم بتجربة ضبط ثلاثة عوامل تدريب ألا وهم`

`A - criterion: gini or entropy` 

`وهو العامل الذي يحدد طريقة الفصل في شجرة القرار`

`B - max_depth`

`أقصى عمق (طول) تصل إليه الشجرة. حيث أن كل ما زاد العمق زاد تعقيد الشجرة`

`C - min_samples_leaf`

`أقل عدد من العينات (الصفوف) في ورقة الشجرة. كل ما زاد عدد العينات في الأوراق كل ما أصبحت الشجرة أبسط `


_____
`سوف نقوم الآن باختبار القيم التالية لعوامل التدريب ورؤية أثرهم على شكل الشجرة وأداؤها
`

`criterion = entropy
max_depth = 3
min_samples_leaf = 3`

```python
custom_model = DecisionTreeClassifier(criterion='entropy', max_depth=3, min_samples_leaf=3, random_state=rs)
```

```python
custom_model.fit(X_train, y_train)
preds = custom_model.predict(X_test)
evaluate_metrics(y_test, preds)
```

`نلاحظ أن دقة النموذج قد تحسنت بشكل طفيف 92.98% ونلاحظ ان قيمة ال إف-ون أيضا تحسنت وأصبحت 90%. والآن فلنقم برسم الشجرة كالتالي`

```python
plot_decision_tree(custom_model, feature_names)
```

`لاحظ كيف أن هذه الشجرة لديها عمق يساوي 3 فقط، وكيف أن الشجرة أصبحت أبسط من السابق وأصبح من السهل تفسيرها. فلنقم الآن بتجربة عوامل مختلفة كالتالي`

`
criterion = gini
max_depth = 10
min_samples_leaf = 3
`

```python
custom_model = DecisionTreeClassifier(criterion='gini', max_depth=10, min_samples_leaf=3, random_state=rs)

custom_model.fit(X_train, y_train)
preds = custom_model.predict(X_test)
display(evaluate_metrics(y_test, preds))

plot_decision_tree(custom_model, feature_names)
```

`نلاحظ الآن أن هذه الشجرة (الشجرة الثالثة) أعقد من الشجرة الثانية وأبسط من الشجرة الأولى. ولكنها الأعلى من حيث الدقة و الإف- ون سكور. الآن ربما تتسآل هل سوف نقوم بتجربة جميع الاحتمالات يدويا من أجل الوصول إلى أفضل شجرة قرار متزنة؟ ولحسن الحظ لا نحتاج إلى ذلك`
